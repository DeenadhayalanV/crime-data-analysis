```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data_frame=pd.DataFrame(pd.read_csv('/content/area_classification.csv'))
```

```python
data_frame.isnull().sum()
```

```python
data_frame['NoOfCrimes'].mean()
```

```python
data_frame.corr()
```

```python
base_color = sb.color_palette()[2]
sb.countplot(data = data_frame, x = 'Class', color = 'lightblue')
```

```python
sorted_counts = data_frame['Level'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, wedgeprops = {'width' : 0.2});
plt.axis('square');
```

```python
base_color = sb.color_palette()[5]
sb.countplot(data = data_frame, y = 'Level', color = base_color)
```

```python
date=data_frame.loc[:,"Date"]
NoOfCrimes=data_frame.loc[:,"NoOfCrimes"]
plt.plot(date,NoOfCrimes)
plt.xlabel('date (hr)')
plt.ylabel('NoOfCrimes (km)')
```

```python
NoOfCrimes=data_frame.loc[:,"NoOfCrimes"]
Area=data_frame.loc[:,"Area/Cluster"]
NoOfCrimes=NoOfCrimes[:10]
Area=Area[:10]
plt.plot(Area,NoOfCrimes)
plt.xlable("Area")
plt.ylabel("NoOfCrimes")
```
